```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shuyangli94/food-com-recipes-and-user-interactions")

print("Path to dataset files:", path)
```

```python
%pip install pandas
```

```python
import pandas as pd
import os

# List all files in the directory
all_files = os.listdir(path)
print("Files in the directory:", all_files)

#Loading all the datasets with pandas
interactions_train = pd.read_csv(os.path.join(path, all_files[0]))
interactions_test = pd.read_csv(os.path.join(path, all_files[1]))
interactions_validation = pd.read_csv(os.path.join(path, all_files[3]))
PP_recipes = pd.read_csv(os.path.join(path, all_files[4]))
RAW_interactions = pd.read_csv(os.path.join(path, all_files[5]))
RAW_recipes = pd.read_csv(os.path.join(path, all_files[6]))
PP_users = pd.read_csv(os.path.join(path, all_files[7]))



```

Measuring missing values

```python
print("Total % of interactions_train missing values: ", (interactions_train.isna().sum().sum()/(interactions_train.shape[0]*interactions_train.shape[1]))*100)
print("Total % of interactions_test missing values: ", (interactions_test.isna().sum().sum()/(interactions_test.shape[0]*interactions_test.shape[1]))*100)
print("Total % of interactions_validation missing values: ", (interactions_validation.isna().sum().sum()/(interactions_validation.shape[0]*interactions_validation.shape[1]))*100)
print("Total % of PP_recipes missing values: ", (PP_recipes.isna().sum().sum()/(PP_recipes.shape[0]*PP_recipes.shape[1]))*100)
print("Total % of RAW_interactions missing values: ", (RAW_interactions.isna().sum().sum()/(RAW_interactions.shape[0]*RAW_interactions.shape[1]))*100)
print("Total % of RAW_recipes missing values: ", (RAW_recipes.isna().sum().sum()/(RAW_recipes.shape[0]*RAW_recipes.shape[1]))*100)
print("Total % of PP_users missing values: ", (PP_users.isna().sum().sum()/(PP_users.shape[0]*PP_users.shape[1]))*100)
```
